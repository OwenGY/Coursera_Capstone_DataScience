```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

Select the first table from the website

```python
pd_table = pd.read_html('http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
for df in pd_table:
    table1=df
    break
```

Rename the table columns and remove the first item, this is the table 

```python
table1.rename(index=str, columns={0: "Postcode", 1: "Borough", 2: "Neighbourhood"}, inplace=True)   
table2 = table1.iloc[1:]
table2.shape
```

```python
table2.head(10)
```

Remove the rows that are not assigned

```python
table2 = table2[table2.Borough != 'Not assigned']
table2
```

For all rows in the table, where Neighbourhood is equal to Not assigned, then neighbourhood is equal to borough

```python
for i in range(0 , len(table2)):
    if table2.Neighbourhood[i] == 'Not assigned':
        table2.Neighbourhood[i] = table2.Borough[i]
        print(table2.iloc[i])
```

```python
data_file = pd.read_csv('https://cocl.us/Geospatial_data/Geospatial_Coordinates.csv')
print(data_file)
```

```python
table2 = table2.reset_index(drop=True)
table2.head(2)
```

```python
data_file.head(2)
data_file.rename(index=str, columns={"Postal Code": "Postcode"}, inplace=True)
data_file.head(2)
```

Add new columns to the table

```python
table2['Latitude']= np.nan
table2['Longitude']= np.nan
table2.head(2)
```

```python
for i in range(0 , len(table2)):
    Latitude1 = 0.0
    Longitude1 = 0.0
    postcode = table2.Postcode[i]
    for p in range (0, len(data_file)):
        if postcode == data_file.Postcode[p]:
            Latitude1 = data_file.Latitude[p]
            Longitude1 = data_file.Longitude[p]
    table2.Latitude[i] = Latitude1
    table2.Longitude[i] = Longitude1
            
table2.head(10)
```

Explore clusters and maps

```python
import folium # map rendering library
import matplotlib.cm as cm
import matplotlib.colors as colors
from geopy.geocoders import Nominatim
print('Imports Completed')
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, borough, neighborhood in zip(table2['Latitude'], table2['Longitude'], table2['Borough'], table2['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
    
map_toronto
```
